```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds
import zipfile as zipfile
import os 
import io
import matplotlib.image as mpimg
```

```
zip_ref = zipfile.ZipFile('/content/archive (77).zip')
zip_ref.extractall('./data/')
zip_ref.close()
```

```
train_dir = '/content/data/archive (77)/train'
test_dir = '/content/data/archive (77)/test'
cats_train_dir = os.path.join(train_dir,'cats')
cats_test_dir = os.path.join(test_dir,'cats')
dogs_train_dir = os.path.join(train_dir,'dogs')
dogs_test_dir = os.path.join(test_dir,'dogs')
```

```
cats_train_files = os.listdir(cats_train_dir)
cats_test_files = os.listdir(cats_test_dir)
dogs_train_files = os.listdir(dogs_train_dir)
dogs_test_files = os.listdir(dogs_test_dir)
```

```
split_size=2
next_cats = [os.path.join(cats_train_dir,fname) for fname in cats_train_files[split_size-2:split_size]]
next_dogs = [os.path.join(dogs_train_dir,fname) for fname in dogs_train_files[split_size-2:split_size]]
```

```
for i,img_path in enumerate(next_cats+next_dogs):
  img = mpimg.imread(img_path)
  plt.imshow(img)
  plt.axis('Off')
  plt.show()
```

```
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```
train_datagen = ImageDataGenerator(rescale=1.0/255.,shear_range=0.2,zoom_range=0.2,width_shift_range=0.2,height_shift_range=0.2,rotation_range=35,horizontal_flip=True,fill_mode='nearest')
test_datagen = ImageDataGenerator(rescale=1.0/255)
```

```
train_data_generator = train_datagen.flow_from_directory(train_dir,target_size=(150,150),class_mode = 'binary')
test_data_generator = test_datagen.flow_from_directory(test_dir,target_size=(150,150),class_mode='binary')
```

```
from tensorflow.keras.applications import InceptionV3
base_model = InceptionV3(input_shape =(150,150,3),include_top=False,weights = 'imagenet')
base_model.summary(())
```

```
need_to_train= False
for layer in base_model.layers:
  if layer.name == 'conv2d_93':
    need_to_train = True
  layer.trainable= need_to_train
```

```
inp = tf.keras.layers.Input(shape=(150,150,3))
x = base_model(inp)
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(64,activation='relu')(x)
x = tf.keras.layers.Dense(128,activation='relu')(x)
x = tf.keras.layers.Dense(1,activation='sigmoid')(x)
model = tf.keras.Model(inputs= inp,outputs=x)
model.compile(optimizer = tf.keras.optimizers.RMSprop(),loss= tf.keras.losses.BinaryCrossentropy(),metrics=['accuracy'])
```

```
history = model.fit(train_data_generator,validation_data=(test_data_generator),epochs = 10,steps_per_epoch=train_data_generator.n//train_data_generator.batch_size,validation_steps=test_data_generator.n/test_data_generator.batch_size)
```
